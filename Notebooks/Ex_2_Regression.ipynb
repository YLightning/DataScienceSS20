# Block 6 Exercise 2: Regression

## Generate Data

```python
#dummy regression data
import numpy as np
from sklearn.datasets import make_regression
X, y = make_regression(n_features=128, n_informative=53,random_state=0, shuffle=False, n_samples=1000)
```

```python
np.shape(X)
```

### E2.1: split data init train and test sets 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

## Train Random Forest Regessor
see https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html#sklearn.ensemble.RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
```

```python
regr = RandomForestRegressor(random_state=0, n_estimators=100,max_depth=10)
```

```python
regr.fit(X_train, y_train)  
```

```python
print(regr.feature_importances_)
```

```python
y_predtest = regr.predict(X_test)
```

```python
y_predtrain = regr.predict(X_train)
```

```python
y_plot = (y_predtest - y_test)*(y_predtest - y_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(y_test)
plt.plot(y_predtest)
```

```python
plt.plot(y_train)
plt.plot(y_predtrain)
```
