## MNIST Data


```python
#imports 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_openml
```

```python
# Load data from https://www.openml.org/d/554
X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
np.shape(X)
```

```python
print(y)
```

### E1.1: split data init train and test sets 


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### E1.2: normalize data

```python
from sklearn.preprocessing import normalize
X_test_n=normalize(X_test, norm="l2", axis=1, copy=True, return_norm=False)
```

```python
X_train_n=normalize(X_train, norm="l2", axis=1, copy=True, return_norm=False)
```

### E1.3: plot some data samples (use https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.imshow.html)

## Random Forest Classification

### E1.5: evaluate three models with differt number of trees: 10, 100, 500
calculate and compare
* accuracy (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html#sklearn.metrics.accuracy_score)
* F-Measure (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html#sklearn.metrics.f1_score)
* plot a ROC curve (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_curve.html#sklearn.metrics.roc_curve)

```python
rfc10 = RandomForestClassifier(n_estimators=10, max_depth=2,random_state=0,n_jobs=4)
rfc10.fit(X_train_n,y_train)
rfc10pred = rfc10.predict(X_test_n)
rfc100= RandomForestClassifier(n_estimators=100, max_depth=2,random_state=0,n_jobs=4)
rfc100.fit(X_train_n,y_train)
rfc100pred = rfc100.predict(X_test_n)
rfc500 = RandomForestClassifier(n_estimators=500, max_depth=2,random_state=0,n_jobs=4)
rfc500.fit(X_train_n,y_train)
rfc500pred = rfc500.predict(X_test_n)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, rfc10pred)
```

```python
accuracy_score(y_test, rfc100pred)
```

```python
accuracy_score(y_test, rfc500pred)
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(y_test, rfc10pred,average="micro")
```

```python
f1_score(y_test, rfc100pred,average="micro")
```

```python
f1_score(y_test, rfc500pred,average="micro")
```

```python
from sklearn.metrics import roc_curve
from sklearn.preprocessing import label_binarize
```

```python
fpr, tpr, thresholds = roc_curve(list(map(int, y_test)), list(map(int, rfc500pred)), pos_label=2)
```
